```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained model and tokenizer
model_name = "gpt2-xl" #openai-community/gpt2-medium #EleutherAI/gpt-neo-2.7B
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
```

```python
def generate_text(prompt, max_length=50, temperature=0.7, top_k=50):
    inputs = tokenizer.encode(prompt, return_tensors="pt")
    outputs = model.generate(
        inputs,
        max_length=max_length,
        temperature=temperature,  # Controls randomness
        top_k=top_k,              # Limits to top-k tokens
        num_return_sequences=1    # Number of texts to generate
    )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
prompt = "what is Full Stack Development?"
generated_text = generate_text(prompt)
print(generated_text)
```
